```python

di={'Emplid':[11,12,3,4,5,6,7],'years_of_service':[10,51,3,5,63,15,20]}
service2=pd.DataFrame(di)
print(service2)
```

```python
employee1.merge(service2)         ## in this some values are common so it is printing it
```

```python
#joins --- 4 types ---1.left join,2.right join,3.inner join,4.outer join


pd.merge(employee1,service2,on="Emplid",how="left")          # it will join common + left 
```

```python
pd.merge(employee1,service2,on="Emplid",how="right")
# it will join common + right
```

```python
pd.merge(employee1,service2,on="Emplid",how="inner")           # it will join only common 
```

```python
pd.merge(employee1,service2,on="Emplid",how="outer")           # it will join common + all 
```
